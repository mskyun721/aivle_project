### 가. LinearRegression 부터 시작해봅시다.

#### [실습문제1] 머신러닝_1
* Train과 Test로 나눈 데이터를 기준으로 LinearRegression 모델링을 진행하고 평가를 해주세요.
* 성능지표 : RMSE, R-squared Score

```python
# 아래에 실습코드를 작성하세요.
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import *
import os

result = {
    'data' : [],
    'model' : [],
    'param' : [],
    'rmse' : [],
    'r2score' : [],
}
```

```python
# 아래에 실습코드를 작성하세요.
line_model = LinearRegression()
line_model.fit(train_x, train_y)
pred = line_model.predict(val_x)

print(mean_squared_error(val_y, pred, squared=False))
print(r2_score(val_y, pred))

result['data'].append(data)
result['model'].append('LinearRegression')
result['param'].append('')
result['rmse'].append(mean_squared_error(val_y, pred, squared=False))
result['r2score'].append(r2_score(val_y, pred))

test_pred = line_model.predict(test_x)
```

```python
# 아래에 실습코드를 작성하세요.
lasso_model = Lasso()
lasso_model.fit(train_x, train_y)
pred = lasso_model.predict(val_x)

print(mean_squared_error(val_y, pred, squared=False))
print(r2_score(val_y, pred))

result['data'].append(data)
result['model'].append('Lasso')
result['param'].append('')
result['rmse'].append(mean_squared_error(val_y, pred, squared=False))
result['r2score'].append(r2_score(val_y, pred))

test_pred = lasso_model.predict(test_x)
```

```python
# 아래에 실습코드를 작성하세요.
Ridge_model = Ridge()
Ridge_model.fit(train_x, train_y)
pred = Ridge_model.predict(val_x)

print(mean_squared_error(val_y, pred, squared=False))
print(r2_score(val_y, pred))

result['data'].append(data)
result['model'].append('Ridge')
result['param'].append('')
result['rmse'].append(mean_squared_error(val_y, pred, squared=False))
result['r2score'].append(r2_score(val_y, pred))

test_pred = Ridge_model.predict(test_x)
```

```python
# 아래에 실습코드를 작성하세요.
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# kears 모델 클리어
keras.backend.clear_session()

# keras 모델 선언
model = keras.models.Sequential()
model.add(Input(shape=(train_x.shape[1])))
model.add(Dense(512))
model.add(Dense(512))
model.add(Dense(256))
model.add(Dense(1))

model.compile(loss='mse'
              , optimizer='adam')

model.summary()
```

```python
# 아래에 실습코드를 작성하세요.
es = EarlyStopping(monitor='val_loss', # 관측 대상
                   min_delta=0,        # 최소한 나빠지지 않으면 괜찮아
                   patience=5,         # 성능 개선되지 않는 걸 얼마나 참을래?
                   verbose=1,
                   restore_best_weights=True # 학습이 멈췄을 때, 최적의 가중치로 전환해줌
                   )
```
